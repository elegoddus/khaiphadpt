# Bước 0: Gộp dữ liệu từ nhiều nguồn
## Đề 13: Phân tích hội thoại trong DVKH trực tuyến
**Mục tiêu:** Gộp 5 file dữ liệu gốc → 1 file `dataset_final.csv` thống nhất

```python
import pandas as pd
import os

# Dataset đã được gộp sẵn từ 5 nguồn:
# - dataset_100_da_dang.txt (100 đoạn)
# - dataset_101_200.txt (100 đoạn)
# - dataset_201_300.txt (100 đoạn)
# - dataset15_5.txt (500 đoạn)
# - Dataset_KPDL_1.xlsx (500 đoạn)

df = pd.read_csv('../data/processed/dataset_final.csv')
print(f'Tổng dòng: {len(df)}')
print(f'Tổng đoạn hội thoại: {df["conv_id"].nunique()}')
print(f'Columns: {list(df.columns)}')
df.head(10)
```

### Thống kê tổng quan

```python
print('=== THỐNG KÊ DATASET ===')
print(f'Tổng đoạn hội thoại : {df["conv_id"].nunique()}')
print(f'Tổng lượt lời       : {len(df)}')
print(f'TB lượt/đoạn        : {len(df)/df["conv_id"].nunique():.1f}')
print()
print('Theo nguồn:')
print(df.groupby('source')['conv_id'].nunique())
print()
print('Theo mức hài lòng:')
print(df.drop_duplicates('conv_id')['final_satisfaction'].value_counts())
print()
print('Theo mức từ lóng:')
print(df.drop_duplicates('conv_id')['slang_level'].value_counts())
print()
print('Top 15 chủ đề:')
print(df.drop_duplicates('conv_id')['topic_name'].value_counts().head(15))
```

### Kiểm tra dữ liệu

```python
print('Missing values:')
print(df.isnull().sum())
print()
print('Dtypes:')
print(df.dtypes)
print()
# Mẫu 1 đoạn hội thoại
sample_conv = df[df['conv_id'] == 1]
for _, row in sample_conv.iterrows():
    label = '[KH]' if row['speaker'] == 'customer' else '[TV]'
    print(f"{label} {row['message']}")
```
